```python
import pandas as pd

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import getpass
```

```python
cl_id = getpass.getpass('Enter client_id: ')
cl_secret = getpass.getpass('Enter client_secret: ')
```

```python
sp = spotipy.Spotify(auth_manager = SpotifyClientCredentials(client_id = cl_id,
                                                           client_secret = cl_secret))
```

### Get a big playlist

```python
playlist = sp.user_playlist_tracks("spotify", "4rnleEAOdmFAbRcNCgZMpY")
```

```python
playlist["total"] 
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3))
    return tracks
```

```python
all_tracks = get_playlist_tracks("4rnleEAOdmFAbRcNCgZMpY")
len(all_tracks)
```

```python
#all_tracks
```

```python
artist_list = []
album_list = []
track_list = []
audio_features_list = []


for track in all_tracks:
    a_list = [name['name'] for name in track['track']['artists']]
    artists = ', '.join(a_list)
    artist_list.append(artists)
    album_list.append(track['track']['album']['name'])
    track_list.append(track['track']['name'])
    audio_features_list.append(sp.audio_features(track['track']['uri'])[0])
    
```

```python
df1 = pd.DataFrame({'artist':artist_list, 'song':track_list, 'album':album_list, 'audio_features_list': audio_features_list})
```

```python
df1
```

```python
dict(df1['audio_features_list'])
pd.DataFrame(dict(df1['audio_features_list']))
df2 = pd.DataFrame(dict(df1['audio_features_list'])).transpose()
```

```python
df2
```

```python
data = pd.concat([df1, df2], axis=1)
```

```python
data = data.drop('audio_features_list', axis=1)
```
